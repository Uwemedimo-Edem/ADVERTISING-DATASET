```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
advert=pd.read_csv("advertising.csv",sep =',')
advert
```

```python
advert.describe()
```

```python
advert.info()
```

```python
#distribution for tv to sales
plt.figure(figsize=(4,2))
sns.histplot(advert['Sales'], bins=10, kde=True,color='MAROON')
plt.title('Advertising distribution')
plt.xlabel('Tv')
plt.ylabel('Sales')
plt.show()
```

```python
#DISTRIBUTION OF NEWSPAPER TO SALES
plt.figure(figsize=(4,2))
sns.histplot(advert['Sales'] , bins =12, kde=True,color='ORANGE')
plt. title('Advertising distribution')
plt.xlabel('Newspaper')
plt.ylabel('Sales')
plt.show()
```

```python
#DISTRIBUTION OF NEWSPAPER TO SALES
plt.figure(figsize=(4,2))
sns.histplot(advert['Sales'] , bins =10,kde=True,color='forestgreen')
plt. title('Advertising distribution')
plt.xlabel('Radio')
plt.ylabel('Sales')
plt.show()
```

```python
#relationship between tv and radio
plt.figure(figsize=(4,2))
sns.histplot(data=advert[['Newspaper','Sales']])
plt.title('Tv vs Sales')
plt.show()
```

```python
sns.boxplot(data=advert[['Newspaper','Radio','Sales']])
plt.title('Boxplot of Radio, Newpaper and Sales')
```

```python
#correlation between newspaper
sns.boxplot(data=advert[['Newspaper']])
plt.title('Newspaper')
plt.show()
```

```python
#heatmap
sns.heatmap(advert.corr(),annot=True,cmap='plasma')
```

```python
#scatterplot of radio vs sales
plt.figure(figsize=(8,4))
plt.scatter(advert['Radio'],advert['Sales'],color='orange');
plt.xlabel('Radio Advertising Budget')
plt.ylabel('Sales')
plt.title('Radio vs. Sales')
plt.show()
```

```python
#scatterplot of radio vs sales
plt.figure(figsize=(8,4))
plt.scatter(advert['TV'],advert['Sales'],color='GREEN');
plt.xlabel('TV Advertising Budget')
plt.ylabel('Sales')
plt.title('TV vs. Sales')
plt.show()
```

```python
#scatterplot of radio vs sales
plt.figure(figsize=(8,4))
plt.scatter(advert['Newspaper'],advert['Sales'],color='purple');
plt.xlabel('Newspaper Advertising Budget')
plt.ylabel('Sales')
plt.title('Newspaper vs. Sales')
plt.show()
```

```python
#correlation between Tv
sns.boxplot(data=advert[['TV']])
plt.title('Tv')
plt.show()
```
